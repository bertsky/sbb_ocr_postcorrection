```python
import pandas as pd
```

```python
df = pd.read_csv('report.csv').drop(columns=["Unnamed: 0"])
```

```python
df.set_index('gt', drop=True).sort_index()
```

```python
df_ocr = df.loc[~df.ocr.str.contains('anahash')]
df_ocr_corrected = df.loc[df.ocr.str.contains('anahash')]
```

```python
print("CER: {} +- {}".format(df_ocr.cer.mean(), df_ocr.cer.std()))
print("WER: {} +- {}".format(df_ocr.wer.mean(), df_ocr.wer.std()))
```

```python
print("CER: {} +- {}".format(df_ocr_corrected.cer.mean(), df_ocr_corrected.cer.std()))
print("WER: {} +- {}".format(df_ocr_corrected.wer.mean(), df_ocr_corrected.wer.std()))
```
