```python
import pandas as pd
from IPython.core.display import display, HTML
```

```python
df = pd.read_csv('report.csv').drop(columns=["Unnamed: 0"])
```

```python
def make_clickable(url, name):
    return '<a href="{}" rel="noopener noreferrer" target="_blank">{}</a>'.format(url,name)

df['ocr'] = df.apply(lambda x: make_clickable("https://qurator-data.de/~kai.labusch/anahash/" + x.report.replace('.json', '.html'), x['ocr']), axis=1)
```

```python
df=df.drop(columns=['report']).set_index('gt', drop=True).sort_index()
```

```python
display(HTML(df.to_html(escape=False)))
```

```python
df_ocr = df.loc[~df.ocr.str.contains('anahash-OCR')]
df_ocr_corrected = df.loc[df.ocr.str.contains('anahash-OCR')]
```

```python
print("CER: {} +- {}".format(df_ocr.cer.mean(), df_ocr.cer.std()))
print("WER: {} +- {}".format(df_ocr.wer.mean(), df_ocr.wer.std()))
```

```python
print("CER: {} +- {}".format(df_ocr_corrected.cer.mean(), df_ocr_corrected.cer.std()))
print("WER: {} +- {}".format(df_ocr_corrected.wer.mean(), df_ocr_corrected.wer.std()))
```
